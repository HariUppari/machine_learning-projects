```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
sharukh = cv2.imread('sharukh.png')
```

```python
sharukh = cv2.cvtColor(sharukh, cv2.COLOR_BGR2RGB)
sharukh = cv2.resize(sharukh , (300,360))
```

```python
sharukh.shape
```

```python
plt.imshow(sharukh)
plt.show()
```

```python
x = sharukh.reshape((-1,3))
print(x.shape)
```

```python
k = 5
model = KMeans(n_clusters = k)
model.fit(x)
```

```python
centroids = model.cluster_centers_
print(centroids)
```

```python
colors = np.array(centroids , dtype = 'uint8')
```

```python
#color patch of size (100, 100 , 3)
i = 1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis('off')
    i = i+1
    mat = np.zeros((100,100,3),dtype = 'uint8')
    mat[:,:,:] = color
    plt.imshow(mat)
    plt.axis('off')
plt.show()
```

```python
model.labels_.shape
```

```python
newimg = np.zeros((x.shape),dtype ='uint8')
newimg.shape
```

```python
for i in range(newimg.shape[0]):
    newimg[i] = colors[model.labels_[i]]
```

```python
sharukhnewimg = newimg.reshape((360, 300, 3))
plt.imshow(sharukhnewimg)
plt.show()
```
